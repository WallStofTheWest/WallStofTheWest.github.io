```python
##Function to calculate Year of Year.

def CalcYoY(date):
    try:
        current_date = date
        current_year = int(date.split('-')[0])
        current_month = date.split('-')[1]

        last_year = current_year - 1
        last_month = current_month

        last_year_date = str(last_year) + "-" + str(last_month)
        print(current_date)
        print(last_year_date)
    #     print(df2[df2.Year_Month == current_date]["Adjusted close"])
    #     print(df2[df2.Year_Month == last_year_date]["Adjusted close"])
        current_value = df2[df2.Year_Month == current_date]["Adjusted close"].values[0]
        last_year_value = df2[df2.Year_Month == last_year_date]["Adjusted close"].values[0]
    #     print(type(current_value))
    #     print(last_year_value)
        YOY = (current_value-last_year_value)/last_year_value
        print(YOY)
        return YOY
    except:
        return "NaN"
    
```
